###Load Dataset from HG

####Booksum Dataset

```python
from datasets import load_dataset

ds = load_dataset("kmfoda/booksum")

import os

# Create a directory in your Google Drive for the dataset
drive_path = './data/Booksum'
os.makedirs(drive_path, exist_ok=True)

# Save each split as a CSV file
for split in ds.keys():
  ds[split].to_csv(f"{drive_path}/{split}.csv")
```

####Govreport Dataset


```python
import pandas as pd
import pyarrow
import os
from datasets import load_dataset

# Create the directory if it doesn't exist
os.makedirs("./data/govreport", exist_ok=True)

parquet_files = {'train':["hf://datasets/ccdv/govreport-summarization@refs/convert/parquet/document/train/0000.parquet",
    "hf://datasets/ccdv/govreport-summarization@refs/convert/parquet/document/train/0001.parquet"],

   'validation':["hf://datasets/ccdv/govreport-summarization@refs/convert/parquet/document/validation/0000.parquet"],

    'test':["hf://datasets/ccdv/govreport-summarization@refs/convert/parquet/document/test/0000.parquet"]
}


ds = load_dataset("parquet", data_files=parquet_files)

# Convert to dataframe and save as csv
for split in parquet_files.keys():
  ds['split'].to_pandas().to_csv(f"./data/govreport/{split}.csv", index=False)
```
